Если длина списка - нечетная, последний элемент остается на своем месте. На экран нужно вывести 2 строчки: арифметическую прогрессию до и после изменений.

**Вход**<br>

```python
# Ex.12
n=int(input('Введите длину списка арифм. прогрессии: '))
x=[1]
for i in range(2,n+1):
    x.append(i)
x_new=x.copy()
for i in range(2,len(x_new)+1,2):
    x_new[i-2]=i
for i in range(1,len(x_new),2):
    x_new[i]=i

print(x,'\n',x_new)
```

### Задача 13 Снежинка (2 балла)
Пользователь вводит нечетное число n. Создайте двумерный массив из n×n элементов, заполнив его символами "." (каждый элемент массива является строкой из одного символа). Затем заполните символами "*" среднюю строку массива, средний столбец массива, главную диагональ и побочную диагональ. <br>
В результате единицы в массиве должны образовывать изображение звездочки. Выведите полученный массив на экран, разделяя элементы массива пробелами.

**Вход**<br>
5<br>
**Выход**<br>
<pre class="output">* . * . *
. * * * .
* * * * *
. * * * .
* . * . *</pre>

**Вход**<br>
3<br>
**Выход**<br>
<pre class="output">* * *
* * *
* * *
</pre>

**Вход**<br>
1<br>
**Выход**<br>
<pre class="output">*
</pre>

```python
# Ex. 13
n=int(input('Введите нечетное n: '))

x=[]
c=[]
a=0
for i in range(n):
    x.append([])
    for j in range(n):
        x[i].append('.')
for i in range(n):
    c.append('*')
    x[n//2][i]=c[i]
    x[i][n//2]=c[i]
    x[i][i]='*'
    x[a][n-(a+1)]='*'
    a+=1
for i in x:
    print(" ".join(i))
```
